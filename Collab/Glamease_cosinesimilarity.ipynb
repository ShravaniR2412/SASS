## Data cleaning

Clean the 'price' column by removing non-numeric characters, converting it to numeric type, and handling missing values.



```python
import re

# Remove non-numeric characters and convert to numeric
df['price'] = df['price'].astype(str).apply(lambda x: re.sub(r'[^\d.]', '', x))
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# Handle missing values (NaN)
# Rationale: Filling NaN with the median is less sensitive to outliers than the mean.
median_price = df['price'].median()
df['price'].fillna(median_price, inplace=True)

display(df.head())
```

## Data preparation

### Subtask:

**Reasoning**:

```python
# Combine 'description' and 'notable_effects' into a new 'text' column
df['text'] = df['description'].fillna('') + ' ' + df['notable_effects'].fillna('')

# Convert text to lowercase
df['text'] = df['text'].str.lower()

display(df.head())
```

## Feature engineering

### Subtask:
Create TF-IDF vectors for the combined text data.


**Reasoning**:
Create TF-IDF vectors for the combined text data using TfidfVectorizer.



```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(max_features=5000)  # Limit vocabulary size
tfidf_matrix = tfidf_vectorizer.fit_transform(df['text'])
```

## Data analysis

### Subtask:
Create a function that filters the dataset based on user input and calculates cosine similarity.


**Reasoning**:
Create the `recommend_products` function as specified in the instructions, incorporating filtering, TF-IDF transformation, cosine similarity calculation, and sorting.



```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def recommend_products(product_type, skintype, min_price, max_price):
    # Filter the DataFrame based on user input
    filtered_df = df[
        (df['product_type'] == product_type) &
        (df['skintype'].str.contains(skintype, na=False)) &
        (df['price'] >= min_price) &
        (df['price'] <= max_price)
    ]

    if filtered_df.empty:
        return []  # Return an empty list if no products match the criteria

    # Transform the filtered data's 'text' column using the pre-fitted tfidf_vectorizer
    filtered_tfidf_matrix = tfidf_vectorizer.transform(filtered_df['text'])

    # Placeholder user input for description
    user_input_text = f"{product_type} {skintype} "  # Combine product type and skin type

    # Transform the user input text using the same vectorizer
    user_tfidf_vector = tfidf_vectorizer.transform([user_input_text])

    # Calculate cosine similarity
    cosine_similarities = cosine_similarity(user_tfidf_vector, filtered_tfidf_matrix).flatten()

    # Sort products by cosine similarity
    sorted_indices = np.argsort(cosine_similarities)[::-1]

    # Get the top 10 recommendations
    top_recommendations = []
    for i in sorted_indices[:10]:
        product_info = {
            'similarity_score': cosine_similarities[i],
            'product_name': filtered_df.iloc[i]['product_name'],
            'price': filtered_df.iloc[i]['price'],
            'picture_src': filtered_df.iloc[i]['picture_src'],
        }
        top_recommendations.append(product_info)

    return top_recommendations



# Call the function with user inputs
recommendations = recommend_products(
    product_type="Face Wash",
    skintype="Oily",
    min_price=50,
    max_price=200
)

# Convert the list of recommendations to a pandas DataFrame
recommendations_df = pd.DataFrame(recommendations)

# Display the DataFrame
display(recommendations_df)
```

```python
import pickle
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Load the pickled TF-IDF vectorizer and DataFrame
with open('tfidf_vectorizer.pkl', 'rb') as f:
    tfidf_vectorizer = pickle.load(f)

with open('recommendation_df.pkl', 'rb') as f:
    df = pickle.load(f)

# (Ensure that your DataFrame has a 'text' column which is the combined text of 'description' and 'notable_effects')
# If not, you can create it as follows:
if 'text' not in df.columns:
    df['text'] = (df['description'].fillna('') + ' ' + df['notable_effects'].fillna('')).str.lower()

def recommend_products(product_type, skintype, min_price, max_price, top_n=10):
    # Filter the DataFrame based on user input
    filtered_df = df[
        (df['product_type'].str.lower() == product_type.lower()) &
        (df['skintype'].str.lower().str.contains(skintype.lower())) &
        (df['price'] >= min_price) &
        (df['price'] <= max_price)
    ]

    if filtered_df.empty:
        return []

    # Transform the filtered products' text column using the loaded vectorizer
    filtered_tfidf_matrix = tfidf_vectorizer.transform(filtered_df['text'])

    # Create a query using user inputs (here we use product_type and skintype; you can expand if needed)
    user_input_text = f"{product_type} {skintype}"
    user_tfidf_vector = tfidf_vectorizer.transform([user_input_text])

    # Calculate cosine similarity between the user's input and each product in the filtered DataFrame
    cosine_similarities = cosine_similarity(user_tfidf_vector, filtered_tfidf_matrix).flatten()

    # Sort the indices of products by similarity score (highest first)
    sorted_indices = np.argsort(cosine_similarities)[::-1]

    # Build a list of the top recommendations
    recommendations = []
    for i in sorted_indices[:top_n]:
        rec = {
            'product_name': filtered_df.iloc[i]['product_name'],
            'price': filtered_df.iloc[i]['price'],
            'picture_src': filtered_df.iloc[i]['picture_src'],
            'similarity_score': cosine_similarities[i]
        }
        recommendations.append(rec)

    return recommendations

# ---------------------
# Take user input
# ---------------------
product_type_input = input("Enter product type (e.g., Face Wash): ")
skintype_input = input("Enter your skin type (e.g., Oily): ")
min_price_input = float(input("Enter minimum price: "))
max_price_input = float(input("Enter maximum price: "))

# Get recommendations based on the user input
results = recommend_products(product_type_input, skintype_input, min_price_input, max_price_input)

# Display the recommendations
if results:
    print("\nTop 10 Recommended Products:")
    for idx, rec in enumerate(results, start=1):
        print(f"Rank {idx}:")
        print(f"  Product Name: {rec['product_name']}")
        print(f"  Price: {rec['price']}")
        print(f"  Image URL: {rec['picture_src']}")
        print(f"  Similarity Score: {rec['similarity_score']:.3f}\n")
else:
    print("No matching products found for your input.")
```
